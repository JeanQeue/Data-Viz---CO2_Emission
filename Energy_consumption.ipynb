## Introduction

### Back to basics, what is CO2? 

Chemically, carbon dioxide is incredibly simple—it is just one carbon atom linked with two oxygen atoms. Together they create a colorless gas that makes up about 0.04% of the greenhouse gases in the Earth's atmosphere.

### So what are Greenhouse gases and Greenhouse effect
Greenhouse gases are gases in an atmosphere that absorb and emit radiation within the thermal infrared (heat) range. This process is the fundamental cause of the greenhouse effect which occurs when greenhouse gasses reflects the radiation that Earth emits, preventing it from escaping into space.

Common greenhouse gases in the Earth's atmosphere include water vapor, carbon dioxide, methane, nitrous oxide, and ozone.

![Greenhouse effect](https://cdn.britannica.com/45/91945-034-DD602A97.jpg)

### Key Facts about Carbon Dioxide Emission:
While carbon dioxide is a natural and necessary component of the atmosphere, the enhanced concentration resulting from human activities uch as burning fossil fuels, deforestation, and industrial processes have raised concerns about its impact on the Earth's climate and ecosystems. Source: https://phys.org/news/2023-02-carbon-dioxide-global.html

1. Human-Induced Impact: The majority of CO2 emissions result from human activities, particularly the combustion of fossil fuels for energy production and transportation.

2. Global Warming Driver: Carbon dioxide is a major contributor to the greenhouse effect, trapping heat in the Earth's atmosphere and leading to a rise in global temperatures.

3. Long-Term Consequences: Excessive CO2 levels can lead to detrimental consequences such as sea level rise, extreme weather events, and disruptions to ecosystems and biodiversity.

4. International Concern: Addressing carbon dioxide emissions is a global concern, with international agreements and initiatives aimed at mitigating the impact of climate change.

### CO2 as an Indicator ?
Carbon dioxide (CO2) emission is a critical environmental indicator that quantifies the release of carbon dioxide gas into the atmosphere, primarily through human activities. As a major greenhouse gas, excessive CO2 emissions contribute significantly to global warming and climate change, posing severe threats to ecosystems, weather patterns, and overall planetary health.

### Line Graphs - CO2 Emissions trend & Per Capita Emissions

```python
# Grouping by year and summing the total emissions for each year
dfx_yearly = dfx.groupby('year')['total'].sum()

# Creating the plot
plt.figure(figsize=(12,8))
plt.plot(dfx_yearly.index, dfx_yearly, marker='o')

# Adding the threshold line
threshold = 50000  # Set your threshold value here
plt.axhline(y=threshold, color='r', linestyle='--', label=f'Threshold ({threshold})')

# Adding the trend line
slope, intercept, r_value, p_value, std_err = stats.linregress(dfx_yearly.index, dfx_yearly)
plt.plot(dfx_yearly.index, intercept + slope * dfx_yearly.index, 'g--', label='Trend line')
plt.title('Total CO2 Emissions by Year')
plt.xlabel('Year')
plt.ylabel('Total CO2 Emissions (metric tons)')
plt.legend()
plt.grid(True)
plt.show()
```

- The total CO2 emissions by year shows a general upward trend in global carbon dioxide emissions from 1975 to 2020 (With some gaps as there were no recorded data for some years)
- The red dashed line represents the trend line, indicating a positive correlation and suggesting a steady increase in emissions over the years.
-  The green dashed line shows the trend of increasing CO2 emissions over time.

```python
# top 10 emitting countries
top_10_emitters = dfx[dfx['country'] != 'World'].groupby('country')['total'].sum().sort_values(ascending=False)[:10].index.tolist()

# Filtering dfx for only the top 10 emitting countries
dfx_top_10 = dfx[dfx['country'].isin(top_10_emitters)]

# Plotting the emission trend for each country
plt.figure(figsize=(12,8))
for country in top_10_emitters:
    country_data = dfx_top_10[dfx_top_10['country'] == country]
    plt.plot(country_data['year'], country_data['total'], marker='o', label=country)
plt.title('CO2 Emission Trends for Top 10 Emitting Countries')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions')
plt.legend()
plt.grid(True)
plt.show()
```

### Bar Graphs - Top & Bottom Emitting countries

```python
# Calculating the total emissions for each country
total_emissions = dfx.groupby('country')['total'].sum().sort_values(ascending=False)

# Geting the top 10 and bottom 10 emitting countries
top_10_emitters = total_emissions[:10]
bottom_10_emitters = total_emissions[-10:]

top_colors = plt.cm.viridis(np.linspace(0, 1, len(top_10_emitters)))
bottom_colors = plt.cm.viridis(np.linspace(0, 1, len(bottom_10_emitters)))

# Bar graph for top 10 emitting countries
plt.figure(figsize=(12,8))
plt.bar(top_10_emitters.index, top_10_emitters, color=top_colors)
plt.title('Top 10 Emitting Countries')
plt.xlabel('Country')
plt.ylabel('Total CO2 Emissions (metric tons)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Bar graph for bottom 10 emitting countries
plt.figure(figsize=(12,8))
plt.bar(bottom_10_emitters.index, bottom_10_emitters, color=bottom_colors)
plt.title('Bottom 10 Emitting Countries')
plt.xlabel('Country')
plt.ylabel('Total CO2 Emissions (metric tons)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

From these graphs we can see a significant disparity in CO2 emissions between the top 10 and bottom 10 emitting countries. The top 10 countries emit far more CO2 than the bottom 10 which could be due to various factors such as industrialization, population, and economic development. The top 10 emitters are primarily industrialized countries or rapidly developing economies which have large industries and populations that contribute to higher CO2 emissions.

The bottom 10 emitters are typically smaller and less developed countries, which have smaller industries and populations, leading to lower CO2 emissions. From this analysis we can say that countries with more industrial activity, which often corresponds to higher economic output, tend to have higher emissions.

### Stacked Area Chart - Emmissions over time by Source

```python
# Grouping by year and sum the emissions for each source
dfx_yearly = dfx.groupby('year')[['coal_co2', 'flaring_co2', 'gas_co2', 'oil_co2']].sum()

# Creating the plot
plt.figure(figsize=(12,8))
plt.stackplot(dfx_yearly.index, dfx_yearly.transpose(), labels=dfx_yearly.columns)
plt.title('Emissions by Source Over Time')
plt.xlabel('Year')
plt.ylabel('Total CO2 Emissions (metric tons)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

From this stacked area chart we can see the contribution of the different CO2 categories over time. Each color in the chart represents a different category. From this chart, we can see the overall height of the stack increases over time, indicating that total CO2 emissions have increased.

The blue area (representing coal emissions) is the largest, suggesting that coal is the dominant source of CO2 emissions. The red and green areas (representing oil and gas emissions respectively) appear to increase over time, indicating a rise in these categories of emissions. The orange area (representing flaring emissions) is relatively small, suggesting that flaring contributes minimally to total CO2 emissions.

### Pie Chart - Countries with highest sources of emissions

```python
# highest oil CO2 emissions and its total emissions
highest_oil_country = dfx.groupby('country')['oil_co2'].sum().idxmax()
highest_oil_emissions = dfx[dfx['country'] == highest_oil_country]['oil_co2'].sum()

# highest gas CO2 emissions and its total emissions
highest_gas_country = dfx.groupby('country')['gas_co2'].sum().idxmax()
highest_gas_emissions = dfx[dfx['country'] == highest_gas_country]['gas_co2'].sum()

# highest flaring CO2 emissions and its total emissions
highest_flaring_country = dfx.groupby('country')['flaring_co2'].sum().idxmax()
highest_flaring_emissions = dfx[dfx['country'] == highest_flaring_country]['flaring_co2'].sum()

# highest coal CO2 emissions and its total emissions
highest_coal_country = dfx.groupby('country')['coal_co2'].sum().idxmax()
highest_coal_emissions = dfx[dfx['country'] == highest_coal_country]['coal_co2'].sum()

# labels
labels = [f'{highest_oil_country} (Oil_Co2)', f'{highest_gas_country} (Gas_Co2)', f'{highest_flaring_country} (Flaring_Co2)', f'{highest_coal_country} (Coal_Co2)']

# Creating the pie chart
plt.figure(figsize=(10,8))
plt.pie([highest_oil_emissions, highest_gas_emissions, highest_flaring_emissions, highest_coal_emissions], labels=labels, autopct='%1.1f%%')
plt.title('Major Co2 Emission Sources & Countries with the Highest source Emissions')
plt.legend(title="Countries and Sources", loc="upper left")
plt.show()
```

From the Pie chart, we can immediately see the dominance of Certain Sources. The United States and China are the dominant contributors to CO2 emissions in their respective categories. The United States leads in oil, gas, and flaring emissions, while China leads in coal emissions.

- Coal emissions from China make up a significant portion (42.8%) of the emissions depicted in the chart, indicating the heavy reliance on coal as an energy source. 

- The United States contributes 38.2% of the emissions in the chart through oil, highlighting the country’s substantial use of oil.

- Gas emissions from the United States also make up a significant portion (18.5%), but flaring emissions are minimal (0.5%), suggesting that flaring is not a major source of CO2 emissions.

### Scatter Plots - Coal emissions vs Gas emissions & Total emissions vs. Per capita emissions

```python
# Scatter plot of coal emissions vs gas emissions
plt.figure(figsize=(12,8))
sns.scatterplot(data=dfx, x='coal_co2', y='gas_co2')
plt.title('Scatter Plot of Coal Emissions vs Gas Emissions')
plt.xlabel('Coal CO2 Emissions (metric tons)')
plt.ylabel('Gas CO2 Emissions (metric tons)')
plt.show()

# Scatter plot of total emissions vs per capita emissions for a specific country
country = 'France' 
dfx_country = dfx[dfx['country'] == country]
plt.figure(figsize=(12,8))
sns.scatterplot(data=dfx_country, x='total', y='co2_per_capita')
plt.title(f'Scatter Plot of Total Emissions vs Per Capita Emissions for {country}')
plt.xlabel('Total CO2 Emissions (metric tons)')
plt.ylabel('Per Capita CO2 Emissions (metric tons)')
plt.show()
```

The first scatter plot shows the relationship between coal CO2 emissions and gas CO2 emissions. The data points are scattered in a pattern that suggests a non-linear relationship between coal and gas emissions. Initially, as coal emissions increase, gas emissions also increase but after a certain point, the trend becomes dispersed.

The second scatter plot shows the relationship between total CO2 emissions and per capita CO2 emissions for France. The data points are aligned in a way that indicates a strong linear correlation between total and per capita emissions. This suggests that as the total emissions for France increase, the per capita emissions also increase proportionally.

### Stacked Bar Chart - Sources of Top Emitting countries

```python
# total emissions for each source for each country
dfx_grouped = dfx.groupby('country')[['coal_co2', 'flaring_co2', 'gas_co2', 'oil_co2']].sum()

# Select the top 10 countries with the most emissions
top_countries = dfx_grouped.sum(axis=1).nlargest(10).index
dfx_top = dfx_grouped.loc[top_countries]

# Create the stacked bar chart
dfx_top.plot(kind='bar', stacked=True, figsize=(12,8))
plt.title('CO2 Emissions from Different Sources for Top 10 Emitting Countries')
plt.xlabel('Country')
plt.ylabel('CO2 Emissions (metric tons)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

The Stacked Bar Chart displays the proportional contributions of the different CO2 sources to the overall total of the countries contributions. The United States and China are the dominant contributors to CO2 emissions among the top 10 emitting countries.
- For many of these countries, especially the United States and China, coal is a significant source of CO2 emissions.
- Different countries have different dominant sources of emissions. For example, coal is a major source for the United States and China, while oil is a significant contributor for Japan.

### Heatmap of Correlation Matrix

```python
# Selecting the numerical columns
numerical_cols = dfx.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr_matrix = numerical_cols.corr()
plt.figure(figsize=(10,8))
plt.xticks(rotation=45, ha='right')
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.xticks(rotation=45)
plt.show()
```

From the heatmap, we can see that variables that have a correlation coefficient close to 1 or -1 have a high correlation. For instance, 'year' and 'CO2' have a correlation of 0.99, indicating a strong positive relationship.

Conversely, variables that have a correlation coefficient close to 0 have a low correlation. For example, 'cement_co2_per_capita' has relatively low correlations with most other variables.

A negative correlation coefficient indicates that as one variable increases, the other decreases. However, from this heatmap, it appears that there are no significant negative correlations between the variables.
